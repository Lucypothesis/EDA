```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set(color_codes=True)
%matplotlib inline
```

```python
# 데이터 불러오기
df = pd.read_csv("hotel_bookings.csv")
# 데이터 샘플 확인
df.head()
```

```python
# 각 칼럼의 속성 및 결측치 확인
df.info() # info() -> 데이터에 대한 전반적인 정보 나타냄
```

```python
# 각 칼럼의 통계치 확인
df.describe()
```

```python
# 각 컬럼의 왜도 확인 # 왜도 = skewness
# df.skew() -> 숫자형이 아닌 데이터(Resor Hotel)이 포함되어있어서 실패
# 해결방법1. 숫자형 열만 선택
df.select_dtypes(include = 'number').skew()
# 해결방법2. numeric_only = True 매개변수 사용
df.skew(numeric_only = True) 
```

```python
# 각 컬럼의 첨도 확인
df.kurtosis(numeric_only = True) # 첨도 = kurtosis
# => babies 칼럼과 같이 분포가 넓지 않은 경우 값이 매우 높게 나옴
```

```python
# 특정 변수 분포 시각화
sns.displot(df['booking_changes'])
```

```python
# 호텔 구분에 따른 lead_time 분포 차이 시각화
sns.violinplot(x='hotel', y='lead_time', data=df, inner=None, color='.8')
sns.stripplot(x='hotel', y='lead_time', data=df, size=1)
```
