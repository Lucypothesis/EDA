```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# 데이터 불러오기
df = pd.read_csv('50_Startups.csv')

# 데이터 샘플 확인
df
```

```python
# Profit 변수로 기본 가로세로 박스 플롯 시각화

# 세로 박스 플롯
plt.figure(figsize=(8,6))
sns.boxplot(y='Profit', data=df)
plt.show()
```

```python
# 가로 박스 플롯
plt.figure(figsize=(8,2))
sns.boxplot(x='Profit', data=df)
plt.show()
```

```python
# State 구분에 따른 Profit 박스 플롯 시각화

plt.figure(figsize=(8,5))
sns.boxplot(x='State', y='Profit', data=df)
plt.show()
```

```python
# 그냥 내가 해보는 주별 R&D 예산 지출
plt.figure(figsize=(8,5))
sns.boxplot(x='State', y='R&D Spend', data=df)
plt.show()
```

```python
# 평균, 데이터 포인트 포함한 박스 플롯 시각화
sns.boxplot(x='State', y='Profit',
            showmeans=True, # 평균에 세모표시하기
            boxprops={'facecolor':'None'}, # 박스 안 색깔 없애기
            data=df)
sns.stripplot(x='State', y='Profit',
              data=df,
              jitter=True, # 연속형 데이터가 겹치지 않도록 약간의 무작위 편향(jitter)이 추가됨. 이를 통해 데이터의 밀도를 더 잘 이해할 수 있음
              marker='o', # 'o': 원(circle), 's': 사각형(square), '^': 삼각형(upward), 'v': 삼각형(downward), 'd': 다이아몬드(diamond)..
              alpha=0.6, # 데이터 포인트의 투명도. 0~1 사이의 값
              color='black')

plt.show()
```
